# LSTM for KOSPI Forecasting Example

```python
import numpy as np 
import pandas as pd
from keras.models import Sequential
from keras.layers import LSTM,Dense
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

dataPath = 'C:/YONG/m1710/myPydata/'

# https://www.kaggle.com/amarpreetsingh/stock-prediction-lstm-using-keras/notebook 
```

### Data Loading 

```python
# S&P 500 Stock Data
# 
data = pd.read_csv(dataPath + 'kospi_20191003.csv',
                  parse_dates=['date'],
                   index_col=['date'])
print(data.columns)
print(data.tail())

cl = data.kospi

print(cl.shape)
print(cl.head())
```

```python
fig = plt.figure(figsize=(12,4))
ax = fig.add_subplot(111)
ax.plot(cl)
ax.grid()
plt.title('KOSPI - 20161003~ 20191002')
ax.tick_params(axis='both', labelsize=8)
plt.show()
```

### Data Preparation

```python
scl = MinMaxScaler()

#Scale the data
cl = np.array(cl).reshape(cl.shape[0],1)
print(cl.max(), cl.min())
print(cl[:5])
cl = scl.fit_transform(cl)
# print(cl[:5])
# print(scl.inverse_transform(cl)[:5])

cl[(735-5):]
```

```python
# What is MinMax Scaling?

plt.scatter(scl.inverse_transform(cl), cl)
plt.xlabel('Original Values')
plt.ylabel('MinMax Scaled Values')
plt.show()
```

```python
#Create a function to process the data into 7 day look back slices

def processData(data,lb):
    X,Y = [],[]
    for i in range(len(data)-lb-1):
        X.append(data[i:(i+lb),0])
        Y.append(data[(i+lb),0])
    return np.array(X),np.array(Y)

X,y = processData(cl,7)
X_train,X_test = X[:int(X.shape[0]*0.85)],X[int(X.shape[0]*0.85):]
y_train,y_test = y[:int(y.shape[0]*0.85)],y[int(y.shape[0]*0.85):]

print('Train set :', X_train.shape[0], y_train.shape[0])
print('Test set :', X_test.shape[0], y_test.shape[0])

print('\nTraining Set : \n', X_train[:3])
# slides one day
```

### Model Building

```python
# Build the model
model = Sequential()
model.add(LSTM(256,input_shape=(7,1), recurrent_dropout=0.4))
model.add(Dense(1))
model.compile(optimizer='adam',loss='mse')

#Reshape data for (Sample,Timestep,Features) 
X_train = X_train.reshape((X_train.shape[0],X_train.shape[1],1))
X_test = X_test.reshape((X_test.shape[0],X_test.shape[1],1))

#Fit model with history to check for overfitting
history = model.fit(X_train,y_train,epochs=300,validation_data=(X_test,y_test),shuffle=False)
# history = model.fit(X_train,y_train,epochs=300)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.legend(['loss', 'validation loss'], loc='upper left')
plt.show()
```

### Assessment of Model Generated Predictions

```python
#      
#    

def PredActPlot(x, y, title):
    plt.plot(scl.inverse_transform(y.reshape(-1,1)))
    Xt = model.predict(x)
    plt.plot(scl.inverse_transform(Xt))
    plt.legend(['Actual', 'LSTM Predicted'], loc='upper left')
    plt.title(title)
    plt.show()

    print('MAE : ' , abs(scl.inverse_transform(y.reshape(-1,1)) - scl.inverse_transform(Xt)).mean())
    

PredActPlot(X_train, y_train, 'Back Test using Train set')    

PredActPlot(X_test, y_test, 'Test against Test set')   
```

```python
Xt = model.predict(X_test)
# print(Xt.shape)
plt.plot(scl.inverse_transform(Xt)[1:] - scl.inverse_transform(y_test.reshape(-1,1))[1:])
plt.plot(scl.inverse_transform(Xt)[1:] - scl.inverse_transform(y_test.reshape(-1,1))[:109])
plt.legend(['Error from the Actual', 'Error from the last day'])
plt.axhline(0, linestyle='--')
plt.show()
```
